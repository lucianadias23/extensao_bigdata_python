```python
grupos = df_pacientes.groupby('ESPECIALIDADE MÉDICA')
grupos
```

# criar nova dataframe com dados especificos

```python
grupo_CirurgiaGeral = grupos.get_group('CIRURGIA GERAL')
```

```python
grupo_CirurgiaGeral.to_excel('grupo_CirurgiaGeral.xlsx', index=False)
```
